```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

def load_data(file_path):
    # Load the CSV file into a Pandas DataFrame
    df = pd.read_csv("C:\\Users\\akhil\\OneDrive\\Documents\\archive[1]\\CAR DETAILS FROM CAR DEKHO.csv")
    return df

def preprocess_data(df):
    # Drop unnecessary columns
    df.drop(columns=['name', 'transmission', 'owner'], inplace=True)
    
    # Create dummy variables for the 'fuel' column
    dummies = pd.get_dummies(df['fuel']).rename(columns=lambda x: 'fuel_' + str(x))
    df = pd.concat([df, dummies], axis=1)
    
    # Drop the original 'fuel' column and one of the dummy columns
    df.drop(columns=['fuel', 'fuel_Diesel'], inplace=True)
    
    # Create dummy variables for the 'seller_type' column
    dummies1 = pd.get_dummies(df['seller_type']).rename(columns=lambda x: 'seller_' + str(x))
    df = pd.concat([df, dummies1], axis=1)
    
    # Drop the original 'seller_type' column and one of the dummy columns
    df.drop(columns=['seller_type', 'seller_Dealer'], inplace=True)
    
    return df

def train_model(X, y):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
    
    # Create a Logistics Regression model
    lr = LogisticRegression()
    
    # Train the model on the training data
    lr.fit(X_train, y_train)
    
    # Make predictions on the testing data
    y_pred = lr.predict(X_test)
    
    # Evaluate the model using accuracy score
    score = accuracy_score(y_pred, y_test)
    print("\nAccuracy score:", score)
    
    return lr

def make_prediction(model, new_data):
    # Make a prediction for a new data point
    prediction = model.predict(new_data)
    print("\nPrediction for new data point:", prediction)

# Load the data
df = load_data('CAR DETAILS FROM CAR DEKHO.csv')

# Preprocess the data
df = preprocess_data(df)

# Split the data into features (X) and target (y)
X = df.iloc[:, [0, 2, 3, 4, 5, 6, 7, 8]].values
y = df.iloc[:, 1].values

# Train the model
model = train_model(X, y)

# Make a prediction for a new data point
new_data = [[2021, 50000, 0, 0, 0, 1, 1, 0]]
make_prediction(model, new_data)
```
